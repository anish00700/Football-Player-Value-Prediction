## 🔧 Data Cleaning & Feature Engineering

```python

df.drop(columns=['Unnamed: 0', 'ID', 'Photo', 'Flag', 'Club Logo'], inplace=True)
df.drop_duplicates(inplace=True)
df.columns = df.columns.str.strip().str.replace(' ', '_').str.lower()
```

```python
df.info()
```

```python

def parse_currency(value):
    if isinstance(value, str):
        value = value.replace('€', '').replace('M', 'e6').replace('K', 'e3')
        try:
            return float(eval(value))
        except:
            return np.nan
    return value

df['value'] = df['value'].apply(parse_currency)
df['wage'] = df['wage'].apply(parse_currency)
df['release_clause'] = df['release_clause'].apply(parse_currency)
```

```python

def height_to_cm(height):
    try:
        feet, inches = height.split("'")
        return round((int(feet) * 12 + int(inches.replace('"', ''))) * 2.54)
    except:
        return np.nan

def weight_to_kg(weight):
    try:
        return round(int(weight.replace('lbs', '')) * 0.453592)
    except:
        return np.nan

df['height_cm'] = df['height'].apply(height_to_cm)
df['weight_kg'] = df['weight'].apply(weight_to_kg)
df.drop(columns=['height', 'weight'], inplace=True)

def age_bucket(age):
    if age <= 21:
        return 'Young'
    elif age <= 28:
        return 'Prime'
    else:
        return 'Veteran'

df['age_group'] = df['age'].apply(age_bucket)
```

Cleaned data and engineered new features like height (cm), weight (kg), and age groups.

##  Univariate Analysis

```python
sns.histplot(df['age'], kde=True, bins=20)
plt.title('Age Distribution')
plt.show()
```

🎯 Most players are aged between 20–30.

```python
sns.histplot(df['overall'], kde=True, bins=20)
plt.title('Overall Rating Distribution')
plt.show()
```

🎯 Ratings cluster around 60–75.

```python
sns.histplot(df['potential'], kde=True, bins=20)
plt.title('Potential Rating Distribution')
plt.show()
```

🎯 Players generally have more potential than current rating.

```python
sns.histplot(df['value'].dropna(), kde=True, bins=30)
plt.xscale('log')
plt.title('Market Value Distribution (Log Scale)')
plt.show()
```

🎯 Value is highly skewed with a few expensive stars.

```python
sns.histplot(df['wage'].dropna(), kde=True, bins=30)
plt.xscale('log')
plt.title('Wage Distribution (Log Scale)')
plt.show()
```

🎯 Wage distribution mirrors market value.

```python
sns.countplot(x='age_group', data=df)
plt.title('Player Count by Age Group')
plt.show()
```

🎯 Most players are in their prime.

##  Bivariate Analysis

```python
sns.scatterplot(x='overall', y='value', data=df, alpha=0.5)
plt.yscale('log')
plt.title('Overall vs Market Value')
plt.show()
```

📌 Higher overall = higher market value.

```python
sns.scatterplot(x='potential', y='value', data=df, alpha=0.5)
plt.yscale('log')
plt.title('Potential vs Market Value')
plt.show()
```

📌 Higher potential generally means higher value.

```python
sns.scatterplot(x='wage', y='value', data=df, alpha=0.5)
plt.xscale('log')
plt.yscale('log')
plt.title('Wage vs Market Value')
plt.show()
```

📌 Strong log-log correlation.

```python
sns.boxplot(x='preferred_foot', y='value', data=df)
plt.yscale('log')
plt.title('Market Value by Preferred Foot')
plt.show()
```

📌 Slight value edge for right-footed players.

```python
sns.boxplot(x='age_group', y='value', data=df)
plt.yscale('log')
plt.title('Market Value by Age Group')
plt.show()
```

📌 Younger players often valued higher.

```python

common_pos = df['position'].value_counts().head(8).index
sns.boxplot(x='position', y='value', data=df[df['position'].isin(common_pos)])
plt.yscale('log')
plt.title('Market Value by Position')
plt.show()
```

📌 Attackers & creators are the most valuable.

##  Multivariate Analysis

```python

plt.figure(figsize=(10, 8))
sns.heatmap(df[['overall', 'potential', 'value', 'wage', 'international_reputation', 'skill_moves']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

📌 `Wage`, `Overall`, and `Potential` are strongly correlated with `Value`.

```python
sns.pairplot(df[['value', 'overall', 'potential', 'wage', 'international_reputation']].dropna())
plt.suptitle('Pairplot of Key Features', y=1.02)
plt.show()
```

📌 Visual confirmation of correlations.

```python
sns.scatterplot(data=df, x='overall', y='value', hue='age_group', alpha=0.6)
plt.yscale('log')
plt.title('Overall vs Value (Hue by Age Group)')
plt.show()
```

📌 Younger players are valued more even at the same rating.

```python
sns.scatterplot(data=df, x='potential', y='wage', hue='preferred_foot', alpha=0.6)
plt.yscale('log')
plt.title('Potential vs Wage (Hue by Foot)')
plt.show()
```

📌 Little difference by preferred foot.

```python
sns.stripplot(data=df, x='international_reputation', y='value', jitter=True, alpha=0.3)
plt.yscale('log')
plt.title('Market Value vs International Reputation')
plt.show()
```

 Reputation has strong influence on market value.

```python
top_clubs = df.groupby('club')['value'].sum().sort_values(ascending=False).head(10)

sns.barplot(x=top_clubs.values, y=top_clubs.index, palette='magma')
plt.title("Top 10 Clubs by Total Market Value")
plt.xlabel("Total Market Value (€)")
plt.ylabel("Club")
plt.tight_layout()
plt.show()
```

```python
top_nations = df.groupby('nationality')['value'].mean().sort_values(ascending=False).head(10)

sns.barplot(x=top_nations.values, y=top_nations.index, palette='viridis')
plt.title("Top 10 Nationalities by Avg Market Value")
plt.xlabel("Average Market Value (€)")
plt.ylabel("Nationality")
plt.tight_layout()
plt.show()
```

```python
# Top 10 most common positions by count
top_positions = df['position'].value_counts().head(10).index

# Average value of players in those positions
avg_value_by_position = df[df['position'].isin(top_positions)].groupby('position')['value'].mean().sort_values(ascending=False)

# Plot
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.barplot(x=avg_value_by_position.values, y=avg_value_by_position.index, palette="crest")
plt.title("Top 10 Positions by Average Market Value")
plt.xlabel("Average Market Value (€)")
plt.ylabel("Position")
plt.tight_layout()
plt.show()
```

We performed a detailed EDA on football player data to identify key factors affecting market value. After cleaning and standardizing features like height, weight, and currency values, we explored patterns using visualizations. Most players are in their prime years with moderate ratings, and market value is highly skewed. Strong correlations were found with wage, overall rating, potential, and international reputation. Younger players tend to be more valuable at similar skill levels.
